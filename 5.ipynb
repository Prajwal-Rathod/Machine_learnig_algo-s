```python
import matplotlib.pyplot as plt
import numpy as np

# Dataset
fruits = ['Banana', 'Orange', 'Apple', 'Grape', 'Pineapple', 'Other']
long = [400, 0, 100, 50, 200, 150]
sweet = [350, 150, 200, 100, 150, 125]
yellow = [450, 300, 50, 30, 100, 25]
total = [500, 300, 250, 150, 250, 200]

# Probabilities
P_fruit = np.array(total) / sum(total)
P_long = sum(long) / sum(total)
P_sweet = sum(sweet) / sum(total)
P_yellow = sum(yellow) / sum(total)

P_long_given_fruit = np.array(long) / np.array(total)
P_sweet_given_fruit = np.array(sweet) / np.array(total)
P_yellow_given_fruit = np.array(yellow) / np.array(total)

# Compute probabilities
P_fruit_given_long_sweet_yellow = P_long_given_fruit * P_sweet_given_fruit * P_yellow_given_fruit * P_fruit
P_fruit_given_long_sweet_yellow /= P_fruit_given_long_sweet_yellow.sum()  # Normalize

# Print results
for fruit, prob in zip(fruits, P_fruit_given_long_sweet_yellow):
    print(f"Probability of {fruit} given Long, Sweet, Yellow: {prob:.4f}")

# Visualization
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12))

# Bar chart of characteristics
bar_width = 0.2
r1 = np.arange(len(fruits))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

ax1.bar(r1, long, color='b', width=bar_width, edgecolor='grey', label='Long')
ax1.bar(r2, sweet, color='g', width=bar_width, edgecolor='grey', label='Sweet')
ax1.bar(r3, yellow, color='r', width=bar_width, edgecolor='grey', label='Yellow')

ax1.set_xlabel('Fruit', fontweight='bold')
ax1.set_xticks([r + bar_width for r in range(len(fruits))])
ax1.set_xticklabels(fruits)

ax1.plot(r1 + bar_width, total, color='k', marker='o', linestyle='-', label='Total')

ax1.set_ylabel('Count')
ax1.set_title('Characteristics of Different Fruits')
ax1.legend()

# Bar chart of probabilities
ax2.bar(fruits, P_fruit_given_long_sweet_yellow, color='c', edgecolor='grey')

ax2.set_xlabel('Fruit', fontweight='bold')
ax2.set_ylabel('Probability')
ax2.set_title('Probability of Fruits Given Long, Sweet, Yellow')

plt.tight_layout()
plt.show()
```
